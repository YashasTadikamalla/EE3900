```python
import numpy as np
import matplotlib.pyplot as plt

def integral(a,x):
    int = 0
    for i in range(len(x)-1):
        int += a[i]*(x[i+1]-x[i])
    return int

W=2
N=1000
f = np.linspace(-2*W,2*W,N)
X_f = np.zeros(N)
M = 100

x_t = np.zeros(M)

for t in range(M):
    for i in range(N):
        if f[i] < -W:
            X_f[i] = 0
        elif f[i] >= W:
            X_f[i] = 0
        else:
            X_f[i] = np.sin(np.pi*f[i])/(np.pi*f[i])*np.e**(2*np.pi*f[i]*((t-(M/2))/20)*1j)
    x_t[t] = integral(X_f,f)

plt.plot(np.linspace(-M/40,M/40,M,endpoint=True),np.abs(x_t))
plt.xlabel("$t$")
plt.ylabel("$|x(t)|$")
plt.grid()
plt.show()
```
