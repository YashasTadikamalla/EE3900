```python
import matplotlib.pyplot as plt
import numpy as np
import math

fs=np.linspace(-2,2,70)

def X(f):
  if (f>2 or f<-2):
    return 0
  else:
    return (math.sin(math.pi*f))/(math.pi*f)

vals_X=[]
vals_S=[]

for i in fs:
  vals_X.append(X(i))
  vals_S.append(X(i)*X(i))

plt.plot(fs,vals_X)
plt.ylabel('X(f)')
plt.xlabel('f')
plt.plot(2, 0,'o',label='Max freq component')
plt.text(2.05,0.05,'W')
plt.plot(-2, 0,'o')
plt.text(-2.05,0.05,'-W')
plt.grid()
plt.legend(loc='upper right')
plt.show()

plt.plot(fs,vals_S)
plt.ylabel('s(f)')
plt.xlabel('f')
plt.plot(2, 0, 'o',label='Max freq component')
plt.text(2.05,0.05,'W')
plt.plot(-2, 0,'o')
plt.text(-2.05,0.05,'-W')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

```python
FS=np.linspace(-8,8,200)

vals_Xs=[]
for i in FS:
  vals_Xs.append(X(i)+X(i-5)+X(i+5)+X(i-10)+X(i+10))

plt.plot(FS,vals_Xs)
plt.ylabel('X_s(f)')
plt.xlabel('f')
plt.plot(2, 0,'o',label='Max freq component')
plt.text(2.05,0.05,'W')
plt.plot(-2, 0,'o')
plt.text(-1.9,0.05,'-W')
plt.plot(5, 0,'o',label='sampling freq')
plt.text(5.05,0.05,'f_s')
plt.plot(3, 0,'o')
plt.text(3.05,0.05,'f_s-W')
plt.plot(7, 0,'o')
plt.text(7.05,0.05,'f_s+W')
plt.plot(-5, 0,'o')
plt.text(-5.05,0.05,'-f_s')
plt.plot(-3, 0,'o')
plt.text(-3.75,0.05,'-f_s+W')
plt.plot(-7, 0,'o')
plt.text(-7.05,0.05,'-f_s-W')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
