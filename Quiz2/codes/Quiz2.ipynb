```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import math

fig, ax = plt.subplots(figsize = (6,6))

ax.set_xlim([-7,7])
ax.set_ylim([-7,7])

pole2, = plt.plot(1/2,0, 'rx',label = 'Pole:z=b')
pole1, = plt.plot(-2,3, 'rx',label = 'Pole:z=a')
zero, = plt.plot(0,0, 'bo',label = 'Zeros')

legend = plt.legend(handles =[pole1,pole2,zero], loc = 'lower right')
fig.gca().add_artist(legend)

circle = plt.Circle([0,0],math.sqrt(13),color = 'w')
fig.gca().add_artist(circle)

circle = plt.Circle([0,0],1,edgecolor = 'black',fill = 0,linestyle = '--')
fig.gca().add_artist(circle)

patches = mpatches.Patch(color="green", label="ROC")
dotted_line = mlines.Line2D([],[],color = 'black',label='Unit circle')
dotted_line.set_linestyle('--')
plt.legend(handles=[patches,dotted_line], loc = 'upper right')
ax.set_facecolor('xkcd:green')

plt.grid()
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("if |a|>1")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import math

fig, ax = plt.subplots(figsize = (6,6))

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])

pole2, = plt.plot(1/4,0, 'rx',label = 'Pole:z=b')
pole1, = plt.plot(-1/4,1/2, 'rx',label = 'Pole:z=a')
zero, = plt.plot(0,0, 'bo',label = 'Zeros')

legend = plt.legend(handles =[pole1,pole2,zero], loc = 'lower right')
fig.gca().add_artist(legend)

circle = plt.Circle([0,0],math.sqrt(5)/4,color = 'w')
fig.gca().add_artist(circle)

circle = plt.Circle([0,0],1,edgecolor = 'black',fill = 0,linestyle = '--')
fig.gca().add_artist(circle)

patches = mpatches.Patch(color="green", label="ROC")
dotted_line = mlines.Line2D([],[],color = 'black',label='Unit circle')
dotted_line.set_linestyle('--')
plt.legend(handles=[patches,dotted_line], loc = 'upper right')
ax.set_facecolor('xkcd:green')

plt.grid()
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("if |a|<1")
plt.show()
```
